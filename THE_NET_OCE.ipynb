```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

# Load the CSV file
file_path = r'C:\Users\Subhadeep Maishal\Music\JUP\test_data\CBN_INPUT.csv'
df = pd.read_csv(file_path, parse_dates=['DATE'])
df.set_index('DATE', inplace=True)  # Set 'DATE' as the index

# List of target columns
target_columns = ['AAO', 'AO', 'DMI', 'MEIv2', 'NAO', 'ONI', 'PDO', 'SOI', 'TNA', 'TSA', 'WHWP', 'WP']

# Create subplots
fig, axes = plt.subplots(nrows=3, ncols=4, figsize=(15, 10))
axes = axes.flatten()

for i, target_column in enumerate(target_columns):
    # Select features and target
    features = df.drop(columns=[target_column])
    target = df[target_column]

    # Split the data into training and testing sets
    X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.2, random_state=42)

    # Standardize the features
    scaler = StandardScaler()
    X_train_scaled = scaler.fit_transform(X_train)
    X_test_scaled = scaler.transform(X_test)

    # Build the ANN model
    model = Sequential()
    model.add(Dense(64, activation='relu', input_dim=X_train.shape[1]))
    model.add(Dense(32, activation='relu'))
    model.add(Dense(1, activation='linear'))  # Output layer for regression

    # Compile the model
    model.compile(optimizer='adam', loss='mean_squared_error')

    # Train the model
    history = model.fit(X_train_scaled, y_train, epochs=50, batch_size=32, validation_split=0.2)

    # Predictions on the test set
    predictions = model.predict(X_test_scaled)

    # Create a DataFrame with actual values, predictions, and DATE column
    result_df = pd.DataFrame({'Actual': y_test.values, 'Predicted': predictions.flatten()}, index=y_test.index)
    result_df['Error'] = result_df['Actual'] - result_df['Predicted']

    # Scatter plot of actual vs predicted values
    axes[i].scatter(result_df['Actual'], result_df['Predicted'], color='blue')
    axes[i].set_title(f'{target_column} - Scatter Plot')
    axes[i].set_xlabel('Actual')
    axes[i].set_ylabel('Predicted')
    axes[i].grid(False)

# Adjust layout
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from scipy.stats import pearsonr

# Load the CSV file
file_path = r'C:\Users\Subhadeep Maishal\Music\JUP\test_data\CBN_INPUT.csv'
df = pd.read_csv(file_path, parse_dates=['DATE'])
df.set_index('DATE', inplace=True)  # Set 'DATE' as the index

# List of target columns
target_columns = ['AAO', 'AO', 'DMI', 'MEIv2', 'NAO', 'ONI', 'PDO', 'SOI', 'TNA', 'TSA', 'WHWP', 'WP']

# Create a DataFrame to store actual and predicted values
result_df = pd.DataFrame(index=df.index)

for target_column in target_columns:
    # Select features and target
    features = df.drop(columns=[target_column])
    target = df[target_column]

    # Split the data into training and testing sets
    X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.2, random_state=42)

    # Standardize the features
    scaler = StandardScaler()
    X_train_scaled = scaler.fit_transform(X_train)
    X_test_scaled = scaler.transform(X_test)

    # Build the ANN model
    model = Sequential()
    model.add(Dense(64, activation='relu', input_dim=X_train.shape[1]))
    model.add(Dense(32, activation='relu'))
    model.add(Dense(1, activation='linear'))  # Output layer for regression

    # Compile the model
    model.compile(optimizer='adam', loss='mean_squared_error')

    # Train the model
    history = model.fit(X_train_scaled, y_train, epochs=50, batch_size=32, validation_split=0.2)

    # Predictions on the test set
    predictions = model.predict(X_test_scaled)

    # Add columns to result_df with aligned indices
    result_df[f'Actual_{target_column}'] = np.nan
    result_df[f'Predicted_{target_column}'] = np.nan
    result_df.loc[y_test.index, f'Actual_{target_column}'] = y_test.values
    result_df.loc[y_test.index, f'Predicted_{target_column}'] = predictions.flatten()

# Calculate correlation coefficients and p-values
corr_actual = result_df[[col for col in result_df.columns if 'Actual' in col]].corr()
corr_predicted = result_df[[col for col in result_df.columns if 'Predicted' in col]].corr()

# Plot separate correlation heatmaps for actual and predicted values with significance indicators
fig, axes = plt.subplots(1, 2, figsize=(18, 8))

# Function to add significance indicators
def add_significance(ax, p_values):
    for i in range(p_values.shape[0]):
        for j in range(p_values.shape[1]):
            if i != j:
                if p_values.iloc[i, j] < 0.05:
                    ax.text(j + 0.5, i + 0.5, '+', ha='center', va='center', fontsize=16, color='red')

# Actual values heatmap
sns.heatmap(corr_actual, annot=True, cmap="coolwarm", linewidths=.5, ax=axes[0])
add_significance(axes[0], corr_actual)
axes[0].set_title('Hit Ocean rapport in Actual Values')

# Predicted values heatmap
sns.heatmap(corr_predicted, annot=True, cmap="coolwarm", linewidths=.5, ax=axes[1])
add_significance(axes[1], corr_predicted)
axes[1].set_title('Hit Ocean rapport in  Predicted Values')

plt.tight_layout()
plt.show()
```
